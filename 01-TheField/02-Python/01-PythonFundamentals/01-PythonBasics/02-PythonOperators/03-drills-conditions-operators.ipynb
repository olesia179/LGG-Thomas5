**11. Write a script that converts euros into dollars.**
- The program will start by asking the user to indicate with a character 'E' or '$' depending on the currency of the amount they are entering.
- Then the program will ask you to enter the amount and display the conversion.

```python
currencies = ('E', 'e', '$')
currency = ""
while currency not in currencies :
    currency = input("Euros into dollars and vise versa converter. Please enter the symbol of currency (E for euros or $ for dollars): ")
amount = float(input("Enter amount please: "))
euro2dollar = 1.08
if currency == "$" :
    print(f"{round(amount/euro2dollar, 2)}E")
else :
    print(f"{round(amount*euro2dollar, 2)}$")
```
