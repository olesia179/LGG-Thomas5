## d) Functions

### Create a Fibonacci function called `fibonacci`  (Fibonacci sequence: 0 1 1 2 3 5 8 13 ...)

It should return the list of all Fibonacci numbers leading up to the Fibonacci number that is bigger (or equal) than the number passed in as argument.
Ex : `fibonacci(10) return [0,1,1,2,3,5,8,13]`
Ex : `fibonacci(5) return [0,1,1,2,3,5]`

````python
def fibonacci(limit) :
    x, y = 0, 1
    result = [x]
    if limit > 0 :
        result.append(y)
    while x + y <= limit :
        y += x
        result.append(y)
        x = y - x
    if y < limit :
        result.append(x + y)
    return result
print(fibonacci(10))
print(fibonacci(5))
print(fibonacci(0))
print(fibonacci(1))
print(fibonacci(2))
print(fibonacci(3))
print(fibonacci(4))
````

### Execute your function. `fibonacci(100)` must be executed in less than 1 second.

````python
import time
start_time = time.time()
print(fibonacci(100))
print(time.time() - start_time)
````

### Create the same function using a generator
Ex : `generate_fibonacci(10) `

````python
def generate_fibonacci(limit) :
    if limit == 0 :
        yield 0
        return
    if limit >= 1 :
        x, y = 0, 1
        yield x
        yield y
        while x + y <= limit :
            y += x
            yield y
            x = y - x
        if y < limit :
            yield x + y
        
print(list(generate_fibonacci(10)))
print(list(generate_fibonacci(5)))
print(list(generate_fibonacci(0)))
print(list(generate_fibonacci(1)))
print(list(generate_fibonacci(2)))
print(list(generate_fibonacci(3)))
print(list(generate_fibonacci(4)))
start_time = time.time()
print(list(generate_fibonacci(100)))
print(time.time() - start_time)
````
